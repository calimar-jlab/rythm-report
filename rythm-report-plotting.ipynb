```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plot
```

```python
%store -r flattened
%store -r pl_scope
```

## plot genre interest over time

```python
# aggregate genres by playlist
genre_over_time_list = []

for df in flattened:
    agg = (
        df.groupby(["playlist_date", "top_genre"])
            .size()
            .reset_index(name = "count")
    )

    genre_over_time_list.append(agg)

genre_over_time = pd.concat(genre_over_time_list, ignore_index = True)

dominant = (
    genre_over_time
        .sort_values(["playlist_date", "count"], ascending = [True, False])
        .groupby("playlist_date", as_index = False)
        .first()
)

plot.figure(figsize = (12, 6))

genres = dominant["top_genre"].unique()

for g in genres:
    subset = dominant[dominant["top_genre"] == g]
    plot.plot(subset["playlist_date"], subset["count"], marker = "o", label = g)

plot.xlabel("Month")
plot.ylabel("Count")
plot.title("Genre Interest Over Time")
plot.legend(title = "Genre")
plot.xticks(rotation = 45)
plot.tight_layout()
plot.show()
```
